### Q11
In a for loop, what is the difference between range(10), range(0, 10), and range(0, 10, 1)?
### A11

```python
# range(10) will return integers from 0 till 9(excluding 10) as default
l = list()
for i in range(10):
    l.append(i)
print(l)

#range(0,10) will return integers in the range [0,10)
l = list()
for i in range(0,10):
    l.append(i)
print(l)

#range(0,10,1) will return integers in the range [0,10) with the jump of 1
l = list()
for i in range(0,10,1):
    l.append(i)
print(l)

#if we want a jump of 2 we can use range(0,10,2)
l = list()
for i in range(0,10,2):
    l.append(i)
print(l)
```
